```python
import matplotlib.pyplot as plt

# Training and validation accuracy values
train_accuracy = [0.5384, 0.5745, 0.6198, 0.6237, 0.6755, 0.6849, 0.6961, 0.7033, 0.7250, 0.7425, 0.7583, 0.7670, 0.7751]
val_accuracy = [0.5103, 0.5840, 0.6309, 0.6477, 0.6540, 0.6677, 0.6752, 0.7039, 0.7333, 0.6608, 0.6771, 0.7458, 0.7601]

# Number of epochs
epochs = range(1, 14)

# Plotting training and validation accuracy
plt.plot(epochs, train_accuracy, label='Training Accuracy')
plt.plot(epochs, val_accuracy, label='Validation Accuracy')

# Adding labels and title
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
#plt.title('Training and Validation Accuracy Over Epochs')

# Adding legend
plt.legend()

# Save the plot as an EPS file
plt.savefig('accuracy_plot.eps', format='eps')

# Display the plot
plt.show()

```
